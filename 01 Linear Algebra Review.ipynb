```python
import matplotlib.pyplot as plt
import numpy as np

# --- Setup the plot ---
fig, ax = plt.subplots(figsize=(10, 10))

# Move the origin to the center of the plot
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.set_xlabel('X-axis', loc='right')
ax.set_ylabel('Y-axis', loc='top')
ax.grid(True, linestyle=':', alpha=1)

# Set limits to see the vectors clearly
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_aspect('equal') # Ensures equal scaling on both axes

# --- Plotting individual vectors ---

# Vector i
vector = [1, 0]
v1 = np.array(vector)
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1 = '+str(vector))

# Vector j
vector = [0, 1]
v2 = np.array(vector)
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2 = '+str(vector))

# Vector 3 = v1 - v2
vector = [-6, -5]
v3 = np.array(vector)
#ax.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='green', label='v3 = '+str(vector))

plt.title('Plotting Vectors')
plt.legend()
plt.show()
```
